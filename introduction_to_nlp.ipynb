```python
synonyms = []#eş anlamlı
antonyms = []#zıt anlamlı

for syn in wordnet.synsets("bad"):
  for l in syn.lemmas():
    #print("l: ",l)
    #print("l.name(): ",l.name())
    synonyms.append(l.name())
    if l.antonyms():
      print("l.antonyms(): ",l.antonyms())
      antonyms.append(l.antonyms()[0].name())

print(set(synonyms))
print(set(antonyms))
```

```python
!pip install nltk

import nltk
nltk.download('movie_reviews')
import pandas as pd
import random

from nltk.corpus import movie_reviews
print(movie_reviews.words())
print(movie_reviews.categories())
print(movie_reviews.fileids())
```

```python
documents =  [(list(movie_reviews.words(fileid)), category)
              for category in movie_reviews.categories()
              for fileid in movie_reviews.fileids(category)]

random.shuffle(documents)
print(documents[1])
```

```python
all_words = []
for w in movie_reviews.words():
  all_words.append(w.lower())

all_words = nltk.FreqDist(all_words)
print(all_words.most_common(4))
print(all_words["stupid"])
```

```python
word_features = list(all_words.keys())[:3000]

def find_features(document):
  words = set(document)
  features = {}

  for w in word_features:
    features[w] = (w in words)

  return features
print(len(movie_reviews.words()))
print(find_features(movie_reviews.words('neg/cv000_29416.txt')))
```

```python
featuresets = [(find_features(rev), category) for (rev, category) in documents]
print(featuresets[0])
```

```python
training_test = featuresets[:4000]
testing_set = featuresets[4000:]

classifier = nltk.NaiveBayesClassifier.train(training_test)
print("Naive Bayes Algo Accuracy percent: ", (nltk.classify.accuracy(classifier, testing_set))*100)
classifier.show_most_informative_features(15)

```
